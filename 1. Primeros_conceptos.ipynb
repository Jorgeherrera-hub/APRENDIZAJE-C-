# Descripcion del archivo "1"

- `#include <iostream>` es una biblioteca de archivos de cabecera que nos permite trabajar con objetos de entrada y salida, como cout

- `using namespace std;` permite ocupar nombres de objetos de las variables de la librería estándar

- `int main()` es una funcion, en donde cualquier codigo que se encuentre dentro de las llaves **{}** se ejecutará

- `cout` es un objeto que se utiliza junto con el operador de inserción (<<) para imprimir texto. (se pronuncia see out)

- `return 0;` termina la funcion

**Cada instrucción de C++ termina con punto y coma ;**

La instrucción `using namespace std;` puede omitirse y sustituirse por la palabra clave `std` seguida del operador `::` para ciertos objetos como `std::cout`

# Tipos de dato 

- `int` - almacena numeros enteros 
- `double` - almacena floats, numeros con decimales 
- `char` - almacena carácteres simples como 'a' o 'B'. Char values se escriben con comillas simples 
- `string` - Almacena texto. String values se escriben con doble comilla
- `bool` - Almacena valores True o False

La sintaxis de creado de variables es

```python

tipo_dato nombre_variable = valor;

```

Es posible solamente declarar el tipo de dato de la variable sin asignarle un valor, este valor tiene que ser asignado dentro de una funcion. **Por ejemplo**


```python

int numero 

int main() {numero = 9 cout<<numero} #El output será 9 


```

También es posible declarar más de una variable al mismo tiempo si es que es el mismo tipo de dato
**Por ejemplo**



```python

int x = 5, y = 6, z = 7



```

# Descripcion del archivo N°5

`cin>>` permite generar un imput para que el usuario escriba algo

se puede almacenar en una variable de la siguiente manera


```python

int main() {int x; cout<<"Escribe un numero: "; cin>> x; } #aquí el numero se guarda en la variable x

```
